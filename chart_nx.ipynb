```python
import pandas as pd 
import matplotlib.pyplot as plt 
import plotly.express as px
import seaborn as sns 
import numpy as np
df = pd.read_csv('netflix_titles.csv')
#ll
```

```python
df.dropna(inplace = True)
df['first_country'] = df['country'].apply(lambda x: x.split(",")[0])
df
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.country = df.country.fillna((df.country).mode()[0])
```

```python
df['count'] = 1
```

```python
df_tv = df[df['type'] == 'TV Show']
df_movie = df[df['type'] == 'Movie']
df['first_country'] = pd.DataFrame(df['country'].apply(lambda x: x.split(',')[0]))
```

```python
df
```

```python

df['first_country'].replace('United States', 'USA', inplace =True)
df['first_country'].replace('United Kingdom', "UK", inplace = True)
df['first_country'].replace('South Korea', 'S.Korea', inplace = True)
```

```python

df_movie_data = df_movie.groupby('first_country')[['first_country','count']].sum().sort_values(by ='count',ascending = False).reset_index()[:10]

df_movie_data = df_movie_data['first_country']
df_movie = df_movie.loc[df_movie['first_country'].isin(df_movie_data)]
df_movie.date_added = pd.to_datetime(df_movie.date_added).dt.year
df_movie = df_movie.groupby('first_country')[['date_added','release_year']].mean().round()

df_movie =df_movie[["date_added","release_year"]].astype(int).sort_values(by = ['date_added'], ascending = True)
df_movie = df_movie.sort_values(by = ['release_year'], ascending = True)
my_range=np.arange(1,len(df_movie.index)+1)
```

```python
def plot(df_movie,x):  
    
    fig,ax = plt.subplots(1,1, figsize = (7,5))

    ax.scatter(df_movie.release_year, my_range, color = "black", s=100, alpha = 0.9, label = "avarage of release year")
    ax.scatter(df_movie.date_added, my_range, color = "#b20710", s=100, alpha = 0.9, label = "average of added year")

    ax.hlines(xmin = df_movie['release_year'], 
              xmax = df_movie['date_added'],
             y = my_range,
             color = "black",
             alpha = 0.2,
             linestyle='dashed')

    ax.yaxis.tick_right()
    plt.yticks(my_range,df_movie.index, fontname = "serif", fontsize = 12)

    fig.text(0.19,0.175, "Released", fontweight = "bold", fontname= "serif", fontsize = 12, color = "black" )
    fig.text(0.80,0.175, "Added", fontweight = "bold", fontname= "serif", fontsize = 12, color = "#b20710" )


    plt.title(x, fontweight= "bold", fontname = "serif", fontsize = 12)

    plt.xlabel('Year')
    plt.ylabel('Country')

```

```python
plot(df_movie,"The age of movie")
```
